```python
r.filter_query('od550aer', ts_type='monthly', flex_ts_type=True, prefer_longer=True)
```

```python
subset = r.filter_files('od550aer', start=2005, stop=2010)
subset
```

```python
monthly = r.filter_files(ts_type='monthly', df=subset)
monthly
```

```python
files = {}

for y in range(2000, 2020):
    try:
        files[y]=r.get_file('od550aer', 'monthly', y)
    except pya.exceptions.DataCoverageError as e:
        files[y] = repr(e)
        
for y, i in files.items():
    print('{}: {}'.format(y, i))
```
