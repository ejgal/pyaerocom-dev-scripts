```python
data = pya.io.ReadUngridded().read('EBASMC', 'concoc')
```

### Get list of unique station names

```python
data.unique_station_names
```

### Get data from Birkenes

```python
birkenes = data.to_station_data('Birken*')
```

### Check variable information (dictionary where keys are variable names and values are metainfo)

```python
birkenes.var_info
```

### Get (pandas) timeseries of OC concentration

```python
tseries = birkenes.concoc
```

```python
tseries.plot();
```

### The StationData object (`birkenes`) has also some plotting and post processing methods

```python
ax = birkenes.plot_timeseries('concoc')
ax = birkenes.plot_timeseries('concoc', 'monthly', ax=ax)
ax = birkenes.plot_timeseries('concoc', 'yearly', marker='o', ls='none', color='g', ms=14, ax=ax)
```
